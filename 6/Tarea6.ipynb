Para el caso de la inserción, el método ``insert`` opera de la siguiente forma:

2- Para un nodo externo (hoja), el método ``insert`` retorna una tupla (lista) de la forma `` [Nodoe(), x, Nodoe()]``

3- Las tuplas siempre son de la forma ``[Nodo izquierdo, valor, Nodo derecho]``.

4- Para un nodo binario, la inserción sea realiza de la siguiente forma:
* Si el valor a insertar es menor que la llave del nodo actual, se inserta recursivamente en el hijo izquierdo.
* Si el valor a insertar es igual a la llave actual, se retorna un puntero al nodo.
* Si el valor a insertar es mayor que la llave del nodo actual, se inserta recursivamente en el hijo derecho.

Ahora, la inserción recursiva puede retornar un puntero a un nodo o una tupla. 
* Si en el nodo actual, la inserción, en alguno de sus hijos, retornó un puntero a un nodo, significa que la inserción ya fue realizada. Por lo tanto, se actualiza el hijo donde se realizó la inserción y se retorna un puntero del nodo actual.
* Si en el nodo actual, la inserción, en alguno de sus hijos,  retornó una tupla, signfica que la inserción no fue realizada y se debe realizar en el nodo actual. Como el nodo actual es un nodo binario, dada la información en la tupla, se puede transformar el nodo binario actual a nodo ternario. La transformación consiste crear dos nuevos hijos y un nuevo campo de información para el nodo actual y mantener uno de sus hijos actuales (dependiendo de donde viene la recursión). Con lo anterior, se tienen todos los elementos para crear un nuevo nodo ternario que reemplaza el nodo binario actual. Finalmente, se retorna el nuevo nodo ternario.

4- Para un nodo ternario, la inserción se realiza de la siguiente forma:
* Si el valor a insertar es menor que la primera llave del nodo actual, se inserta recursivamente en el hijo izquierdo.
* Si el valor a insertar es igual a la primera llave del nodo actual, se retorna un puntero al nodo actual.
* Si el valor a insertar es mayor que la primera llave pero menor que la segunda llave del nodo actual, se inserta recursivamente en el hijo del medio.
* Si el valor a insertar es igual a la segunda llave actual, se retorna un puntero al nodo.
* Si el valor a insertar es mayor que la segunda llave y, por lo tanto, mayor que primera llave del nodo actual, se inserta recursivamente en el hijo del derecho.

Ahora, la búsqueda recursiva puede retornar un puntero a un nodo o una tupla. 
* Si en el nodo actual, la inserción, en alguno de sus hijos, retornó un puntero a un nodo, significa que la inserción ya fue realizada. Por lo tanto, se actualiza el hijo donde se realizó la inserción y se retorna un puntero del nodo actual.

Como el nodo actual es un nodo ternario, no se puede realizar la inserción en este nodo. Por lo tanto, se debe actualizar la información en la tupla y retornar la tupla. La tupla se actualiza eligiendo un nuevo valor como el valor medio entre el valor a insertar y la primera y segunda llave del nodo actual y actualizando el nodo izquierdo de la tupla y el nodo derecho de la tupla con los atributos del nodo actual (dependiendo de donde viene la recursión). Luego, se retorna la tupla actualizada.

5- De esta forma, llamando al método ``insert`` se realiza la inserción recursiva en el árbol, partiendo en la raíz hasta llegar a un nodo externo (hoja). Luego, en la hoja se retornará recursivamente un tupla que, en caso de poder ser insertada, se insertará en un nodo respectivo y, desde ese nodo, se retornará recursivamente un puntero hacia el nodo actual. El método ``insert`` del árbol puede retornar un nodo y, por lo tanto, el nodo es la nueva raíz.
````python
#Clase que representa un nodo con dos hijos y una llave
class Nodo2:

    #Método para inicializar un objeto de la clase.
    def __init__(self, izq, info, der):
        ##Atributos de la clase
        #Hijo izquierdo
        self.izq=izq
        #LLave
        self.info=info
        #Hijo derecho
        self.der=der

    #Método recursivo que, dado un entero x que representa una llave, inserta la llave (en caso de ser necesario) en su lugar correspondiente.
    def insert(self,x):

        #Si la llave x es menor que la llave del nodo actual, se inserta recursivamente en su hijo izquierdo
        if x < self.info:
            node = self.izq.insert(x)

            #En caso que la función recursiva retorne un puntero al hijo izquiedo, significa que la llave ya se insertó en el hijo izquierdo. Por lo tanto, se actualiza el hijo izquierdo (que contiene a la llave insertada x) y se retorna un puntero al nodo actual (actualizado)
            if isinstance(node,Nodo2) or isinstance(node,Nodo3):
                self.izq = node
                return self

            #En caso que la función recursiva no retorne un puntero al hijo izquierdo (retorna una tupla), significa que la llave asociada a la tupla se tiene que insertar ahora. Por lo tanto, se crea un nuevo nodo ternario para así agregar la llave asociada a la tupla en el nodo actual (en su posición respectiva) y retornar un puntero hacia el nodo actual (actualizada).
            else:
                #Se crea un nuevo hijo izquierdo, se crea un hijo del medio y se crean las llaves del nodo ternario.
                izq = node[0]
                info1 = node[1]
                med = node[2]
                info2 = self.info
                #Se mantiene el hijo derecho 
                der = self.der
                #Se crea el nodo ternario y se retorna
                node = Nodo3(izq,info1,med,info2,der)
                return node

        #La llave ya se encuentra en el nodo, luego se retorna el nodo sin modificaciones.
        elif x == self.info:
            return self

        #Si la llave x es mayor que la llave del nodo actual, se inserta recursivamente en su hijo derecho.
        else:
            node = self.der.insert(x)

            #En caso que la función recursiva retorne un puntero al hijo derecho, significa que la llave ya se insertó en el hijo derecho. Por lo tanto, se actualiza el hijo derecho (que contiene a la llave insertada) y se retorna un puntero al nodo actual (actualizado).
            if isinstance(node,Nodo2) or isinstance(node,Nodo3):
                self.der = node
                return self

            #En caso que la función recursiva no retorne un puntero al hijo derecho (retorna una tupla), significa que la llave asociada a la tupla se tiene que insertar ahora. Por lo tanto, se crea un nuevo nodo ternario para así agregar la llave asociada a la tupla en el nodo actual (en su posición respectiva) y retornar un puntero hacia el nodo actual (actualizado).
            else:
                #Se mantiene el hijo izquierdo.
                izq = self.izq
                #Se crea un nuevo hijo derecho, se crea un hijo del medio y se crean las llaves del nodo ternario.
                info1 = self.info
                med = node[0]
                info2 = node[1]
                der = node[2]
                #Se crea el nodo ternario y se retorna
                node = Nodo3(izq,info1,med,info2,der)
                return node

    #Método recursivo que, dado un entero x que representa una llave, se busca la llave en el nodo actual.
    def search(self,x):

        #Si la llave buscada es igual a la llave del nodo actual, significa que se encontró la llave y, por lo tanto, se retorna un puntero al nodo actual.
        if x == self.info:
            return self

        #Si la llave buscada es menor que la llave del nodo actual, se busca recursivamente la llave en el hijo izquierdo.
        elif x < self.info:
            return self.izq.search(x)

        #Si la llave buscada es mayor que la llave del nodo actual, se busca recursivamente la llave en el hijo derecho.
        else:
            return self.der.search(x)

    #Método recursivo que imprime la llave del nodo acutal y las llaves asociadas al hijo izquierdo y derecho.
    def string(self):
        return ("("+self.izq.string()
                +str(self.info)
                +self.der.string()+")")
````
